```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, LabelEncoder
```

```python
label_encoder = LabelEncoder()
df["Pass_Fail"] = label_encoder.fit_transform(df["Pass_Fail"])  # Pass=1, Fail=0
```

```python
binary_cols = ["Internet_Access_at_Home", "Extracurricular_Activities"]
for col in binary_cols:
    df[col] = df[col].map({"Yes": 1, "No": 0})
```

```python
# Encode Gender
df["Gender"] = df["Gender"].map({"Male": 0, "Female": 1})
```

```python
# Split features & target
X = df.drop(columns=["Pass_Fail"])
y = df["Pass_Fail"]

# Encode Parental_Education_Level using one-hot encoding
X = pd.get_dummies(X, columns=['Parental_Education_Level'], drop_first=True)
```

```python
# Normalize features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
# Train-test split
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42, stratify=y)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout
from tensorflow.keras.optimizers import Adam
```

```python
model = Sequential([
    Dense(32, input_dim=X_train.shape[1], activation="relu"),
    Dropout(0.2),
    Dense(16, activation="relu"),
    Dropout(0.2),
    Dense(1, activation="sigmoid")  # Binary classification
])
```

```python
# Compile model
model.compile(
    loss="binary_crossentropy",
    optimizer=Adam(learning_rate=0.001),
    metrics=["accuracy"]
)
```

```python

# 5. Train Model
history = model.fit(
    X_train, y_train,
    epochs=50,
    batch_size=16,
    validation_data=(X_test, y_test),
    verbose=1)
```

```python
# 7. Plot Training History (Optional)
import matplotlib.pyplot as plt

plt.plot(history.history["accuracy"], label="Train Accuracy")
plt.plot(history.history["val_accuracy"], label="Validation Accuracy")
plt.xlabel("Epochs")
plt.ylabel("Accuracy")
plt.legend()
```

```python
y_pred = (model.predict(X_test) > 0.5).astype("int32")

print("Accuracy:", accuracy_score(y_test, y_pred))
print(confusion_matrix(y_test, y_pred))
print(classification_report(y_test, y_pred))
```
