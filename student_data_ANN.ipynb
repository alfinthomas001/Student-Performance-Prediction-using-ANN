```python
# Encoding and Scaling
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, LabelEncoder
```

```python
label_encoder = LabelEncoder()
df["Pass_Fail"] = label_encoder.fit_transform(df["Pass_Fail"])
```

```python
df["Internet_Access_at_Home"] = label_encoder.fit_transform(df["Internet_Access_at_Home"])
```

```python
df["Extracurricular_Activities"] = label_encoder.fit_transform(df["Extracurricular_Activities"])
```

```python
df["Parental_Education_Level"] = label_encoder.fit_transform(df["Parental_Education_Level"])
```

```python
# Encode Gender
df["Gender"] = label_encoder.fit_transform(df['Gender'])
```

```python
# Split features & target
X = df.drop(columns=["Pass_Fail"])
y = df["Pass_Fail"]
```

```python
# Normalize features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
# Train-test split
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42, stratify=y)
```

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
```

```python
lr.fit(X_train,y_train)
```

```python
y_pred=lr.predict(X_test)
```

```python
from sklearn.metrics import mean_absolute_error,r2_score,mean_squared_error
```

```python
print('MAE =',mean_absolute_error(y_test,y_pred))
```

```python
print('MSE =',mean_squared_error(y_test,y_pred))
```

```python
print('r2 =',r2_score(y_test,y_pred))
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dt=DecisionTreeClassifier()
```

```python
dt.fit(X_train,y_train)
```

```python
y_pred=dt.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
```

```python
print('accuracy =', accuracy_score(y_pred,y_test))
print('confusionmatrix =', confusion_matrix(y_pred,y_test))
print('cr =', classification_report(y_pred,y_test))
```

```python
#ANN model
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout
from tensorflow.keras.optimizers import Adam
```

```python
model = Sequential([
    Dense(32, input_dim=X_train.shape[1], activation="relu"),
    Dropout(0.2),
    Dense(16, activation="relu"),
    Dropout(0.2),
    Dense(1, activation="sigmoid")  # Binary classification
])
```

```python
# Compile model
model.compile(loss="binary_crossentropy",optimizer=Adam(learning_rate=0.001),metrics=["accuracy"])
```

```python
#Train Model
history = model.fit(X_train, y_train,epochs=50,batch_size=16,validation_data=(X_test, y_test),verbose=1)
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
```

```python
y_pred = (model.predict(X_test) > 0.5).astype("int32")

print("Accuracy:", accuracy_score(y_test, y_pred))
print(confusion_matrix(y_test, y_pred))
print(classification_report(y_test, y_pred))
```
